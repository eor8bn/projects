```python
import pandas as pd

import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://sbmrne.ru/magazine#section21'
```

```python
page = requests.get(url)
```

```python
soup = bs(page.text, "html.parser")
```

```python
itemCoffee = soup.findAll('div', class_='product-item')
```

```python
specialty_coffee_df = pd.DataFrame(columns=['Roasters', 'Name', 'Subname', 'Treatment', 'Acidity', 'Density', 'Descriptors', 'Weight', 'Price'])
```

```python
for ic in itemCoffee:
    itemCoffee_center = ic.findAll('div', class_="product-item--center")
    for icc in itemCoffee_center:
        fullname = icc.find('div', class_="product-item--name").text[:-1].split('\n')[1:]
        name = fullname[0].strip()
        subname = fullname[1]
        treatment = icc.find('div', class_="product-item--treatment").text.strip()[11:]
        try: 
            quality = icc.findAll('dl', class_="product-item--quality-item")
            acidity = len(quality[0].findAll('span', class_="full"))
            density = len(quality[1].findAll('span', class_="full"))
        except:
            continue
        descriptors = icc.find('div', class_="product-item--taste").text[1:-1]
    itemCoffee_prices = ic.findAll('div', class_="product-item--prices")
    for icp in itemCoffee_prices:
        try:
            weight = int(icp.find('div', class_="col-4").find('span').text[:-2])
        except:
            continue
        price = int(icp.find('div', class_="product-item--prices-item_price").text.strip().replace(' ',''))
    this_coffee_dict = {
        'Roasters': 'SUBMARINE',
        'Name': name,
        'Subname': subname,
        'Treatment': treatment,
        'Acidity': acidity,
        'Density': density,
        'Descriptors': descriptors,
        'Weight': weight,
        'Price': price
    }
    this_coffee_df = pd.DataFrame([this_coffee_dict])
    specialty_coffee_df = pd.concat([specialty_coffee_df, this_coffee_df], ignore_index=True)
```

```python
specialty_coffee_df
```

```python
specialty_coffee_df.drop(specialty_coffee_df[specialty_coffee_df.Weight == 1].index, inplace = True)
```

```python
specialty_coffee_df['Price_per_100'] = specialty_coffee_df.Price / specialty_coffee_df.Weight * 100
```

```python
specialty_coffee_df = specialty_coffee_df.sort_values(by=['Name'])
```

```python
specialty_coffee_df = specialty_coffee_df.reset_index(drop = True)
```
